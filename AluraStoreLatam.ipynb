#1. Análisis de facturación



```python
ingreso_total_tienda_1 = tienda["Precio"].sum()
ingreso_total_tienda_2 = tienda2["Precio"].sum()
ingreso_total_tienda_3 = tienda3["Precio"].sum()
ingreso_total_tienda_4 = tienda4["Precio"].sum()
```

```python
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
ingresos = [ingreso_total_tienda_1, ingreso_total_tienda_2, ingreso_total_tienda_3, ingreso_total_tienda_4]
```

```python
plt.figure(figsize=(8, 5))
plt.bar(tiendas, ingresos,
        color='orange',
        edgecolor='black')

for value in ingresos:
    plt.text(ingresos.index(value) , value + 5, '$'+str(value), ha='center')

plt.title("Ingresos por Tienda")
plt.xlabel("Tienda")
plt.ylabel("Ingreso")

plt.show()
```

```python
dif_may_men = ingreso_total_tienda_1 - ingreso_total_tienda_4
print(f'La diferencia entre la tienda que más vende y la que menos vende es: ${dif_may_men}, siendo la tienda 1: ${ingreso_total_tienda_1} y tienda 4: ${ingreso_total_tienda_4}')
```

```python
porcentaje_de_diferencia_ventas_totales = (dif_may_men / ingreso_total_tienda_1) * 100
print(f'el porcentaje de la firencia de ventas entre las tiendas es: {round(porcentaje_de_diferencia_ventas_totales, 2)}')
```

```python
import folium
import seaborn as sns
```

```python
correlaciones_tienda_1 = tienda[['lat', 'lon', 'Precio']].corr()
```

```python
# Mostrar mapa de calor
plt.figure(figsize=(8, 6))
sns.heatmap(correlaciones_tienda_1, annot=True, cmap='coolwarm')
plt.title('Mapa de Correlación')
plt.show()
```

```python
# Crear el mapa centrado
m = folium.Map(location=[tienda['lat'].mean(), tienda['lon'].mean()], zoom_start=12)

# Agregar puntos al mapa con información
for _, row in tienda.iterrows():
    folium.Marker(
        location=[row['lat'], row['lon']],
        popup=f"Zona: {row['Lugar de Compra']}<br>Ventas: {row['Categoría del Producto']}"
    ).add_to(m)

# Guardar o mostrar mapa
m.save("mapa_por_zonas.html")
```
