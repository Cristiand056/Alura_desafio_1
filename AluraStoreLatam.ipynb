### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_total_tienda_1 = tienda["Precio"].sum()
ingreso_total_tienda_2 = tienda2["Precio"].sum()
ingreso_total_tienda_3 = tienda3["Precio"].sum()
ingreso_total_tienda_4 = tienda4["Precio"].sum()
```

```python
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
ingresos = [ingreso_total_tienda_1, ingreso_total_tienda_2, ingreso_total_tienda_3, ingreso_total_tienda_4]
```

```python
plt.figure(figsize=(8, 5))
plt.bar(tiendas, ingresos,
        color='orange',
        edgecolor='black')

for value in ingresos:
    plt.text(ingresos.index(value) , value + 5, '$'+str(value), ha='center')

plt.title("Ingresos por Tienda")
plt.xlabel("Tienda")
plt.ylabel("Ingreso")

plt.show()
```

```python
dif_may_men = ingreso_total_tienda_1 - ingreso_total_tienda_4
print(f'La diferencia entre la tienda que más vende y la que menos vende es: ${dif_may_men}, siendo la tienda 1: ${ingreso_total_tienda_1} y tienda 4: ${ingreso_total_tienda_4}')
```

```python
porcentaje_de_diferencia_ventas_totales = (dif_may_men / ingreso_total_tienda_1) * 100
print(f'el porcentaje de la firencia de ventas entre las tiendas es: {round(porcentaje_de_diferencia_ventas_totales, 2)}')
```

# 2. Ventas por categoría

```python
venta_por_categorias_1 = tienda['Categoría del Producto'].value_counts().reset_index()
venta_por_categorias_1
```

```python
# Filtrar las 3 categorías con mayor cantidad vendida
top_3_1 = venta_por_categorias_1.nlargest(3, 'count')

# Gráfica
plt.figure(figsize=(8, 5))
plt.bar(top_3_1['Categoría del Producto'],
        top_3_1['count'],
        color='orange',
        edgecolor='black')

# Títulos y etiquetas
plt.title('Top 3 Categorías de Producto Más Vendidas', fontsize=14)
plt.xlabel('Categoría del Producto')
plt.ylabel('Cantidad Vendida')

# Etiquetas encima de las barras
for index, value in enumerate(top_3_1['count']):
    plt.text(index, value + 2, str(value), ha='center')

plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```
